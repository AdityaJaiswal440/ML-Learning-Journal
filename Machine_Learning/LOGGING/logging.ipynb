```python
import logging

logging.basicConfig(level=logging.DEBUG)

logging.debug("This's a debug message")
logging.info("This's an info message")
logging.warning("This is a warning message")
logging.error("This's an error message")
logging.critical("This's a critical message")
```

```python
import logging

logging.basicConfig(
    filename='app.log',
    filemode='a',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True
)

logging.debug("This's a debug message")
logging.info("This's an info message")
logging.warning("This is a warning message")
logging.error("This's an error message")
logging.critical("This's a critical message")
```

```python
# Multiple Loggers

import logging

logger1 = logging.getLogger("module-1")
logger1.setLevel(logging.DEBUG)

logger2 = logging.getLogger("module-2")
logger2.setLevel(logging.WARNING)

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
)
logger1.debug("DEbug msg for module 1")
logger2.warning("Warning msg for module 2")
logger2.error("This's an error msg.")
```
