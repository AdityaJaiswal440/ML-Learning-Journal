```python
class GeneralCustomer:
    def __init__(self,cust_ID,name):
        self.cust_ID=cust_ID
        self.name=name
    
    def get_basic_info(self):
        return f"ID: {self.cust_ID} | Name: {self.name}"

class StreamingCustomer(GeneralCustomer):
    def __init__(self, cust_ID, name,watch_time):
        super().__init__(cust_ID, name)
        self.watch_time = watch_time
        
    def get_streaming_report(self):
        return f"{self.name} has watched {self.watch_time} hours this month"
    
class RetailCustomer(GeneralCustomer):
    def __init__(self, cust_ID, name,total_spent):
        super().__init__(cust_ID, name)
        self.total_spent = total_spent
    
    def get_retail_report(self):
        return f"{self.name} spent ₹{self.total_spent} recently."
    
class VIPCustomer(StreamingCustomer):
    def __init__(self, cust_ID, name, watch_time,concierge_assigned):
        super().__init__(cust_ID, name, watch_time)
        self.concierge_assigned = concierge_assigned
    
    def get_streaming_report(self):
        super().get_streaming_report
        return f"🌟 This is a Priority User | \n{ super().get_streaming_report()}"
    
vip_user = VIPCustomer("VIP_777", "Abhay Sahu", 150, True)

print(vip_user.get_basic_info()) 

print(vip_user.get_streaming_report())
```
