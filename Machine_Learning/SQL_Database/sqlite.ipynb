```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
Create Table If not Exists employees(
    id Integer Primary key,
    name Text Not Null,
    age Integer,
    department text
)
''')

connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
cursor.execute('''
Insert into employees(name,age,department)
Values('Aditya',19,'AI-DS')               
''')

cursor.execute('''
Insert into employees(name,age,department)
Values('Aayush',19,'CS-AI')               
''')

cursor.execute('''
Insert into employees(name,age,department)
Values('Aman',22,'IT-DS')               
''')

connection.commit()
```

```python
cursor.execute('''Select * from employees''')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
Update employees
set age=20
where name='Aditya'           
''')

connection.commit()
```

```python
cursor.execute('''
Delete from employees 
where name='Aman'
''')

connection.commit()
```

```python
cursor.execute('''Select * from employees''')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

cursor.execute('''
Create Table if not exists sales(
    id Integer Primary Key,
    date Text not Null,
    product Text not Null,
    sales Integer,
    region Text
)
''')

sales_data = [
    ('2026-01-1','Product1',100,'North'),
    ('2026-01-2','Product2',200,'South'),
    ('2026-01-3','Product3',300,'East'),
    ('2026-01-4','Product4',200,'West'),
    ('2026-01-5','Product5',400,'North')
]

cursor.executemany('''
Insert into sales(date,product,sales,region)
values(?,?,?,?)
''',sales_data)

connection.commit()
```

```python
cursor.execute('Select * from sales')
rows = cursor.fetchall()

for row in rows:
    print(row)
```
