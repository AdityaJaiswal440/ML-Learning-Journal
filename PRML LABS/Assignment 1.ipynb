15. Use Case: Linear Algebra
Solve the system of linear equations:
3x+2y+z=10
2x+3y+3z=18
x+y+2z=8
Use NumPy&#39;s linalg.solve to find x,y,z.

```python
A = np.array([[3, 2, 1],
              [2, 3, 3],
              [1, 1, 2]])

B = np.array([10, 18, 8])

solution = np.linalg.solve(A, B)
x, y, z = solution
print(f"Results:\nx = {x:.2f}\ny = {y:.2f}\nz = {z:.2f}")
```

1. Line Plot

Create a line plot for the following data:
o X-axis: Numbers from 1 to 10
o Y-axis: Squares of the numbers (e.g., y=x2y = x^2y=x2)
Customize the plot by adding:
o A title
o Labels for the X and Y axes
o A grid

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(1,11)
y=x**2

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2, label='$y = x^2$')

ax.set_title("Line Plot: Relationship between $x$ and $x^2$", fontsize=14, fontweight='bold')
ax.set_xlabel("Input Value ($x$)", fontsize=12)
ax.set_ylabel("Squared Value ($y$)", fontsize=12)

ax.grid(True, linestyle='--', alpha=0.7)
ax.legend()

plt.show()
```

2. Bar Plot
Create a bar plot to visualize the following data:
o Categories: Apples, Bananas, Cherries, Dates
o Values: 40, 25, 35, 20
Add:
o Different colors for each bar
o A legend

```python
data = {
    'Fruit': ['Apples', 'Bananas', 'Cherries', 'Dates'],
    'Value': [40, 25, 35, 20]
}
df = pd.DataFrame(data)

df = df.sort_values(by='Value')

fruit_colors = {
    'Apples': 'red',
    'Bananas': 'gold',
    'Cherries': 'crimson',
    'Dates': 'saddlebrown'
}
colors = [fruit_colors[fruit] for fruit in df['Fruit']]

fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(df['Fruit'], df['Value'], color=colors)

ax.set_xlabel('Fruit Categories', fontsize=12)
ax.set_ylabel('Quantity / Value', fontsize=12)
ax.set_title('Distribution of Fruits (Sorted)', fontsize=14, fontweight='bold')

ax.legend(bars, df['Fruit'], title='Fruit Types', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('bar_plot.png') 
plt.show()
```

4. Subplots
Create a figure with two subplots:
o Subplot 1: A sine wave plot for values of xxx from 000 to 2π2\pi2π
o Subplot 2: A cosine wave plot for values of xxx from 000 to 2π2\pi2π
Ensure both subplots have appropriate titles and axis labels.

```python
x = np.linspace(0, 2 * np.pi, 100)
y_sine = np.sin(x)
y_cosine = np.cos(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x, y_sine, color='royalblue', linewidth=2, label='$\sin(x)$')
ax1.set_title("Sine Wave Analysis", fontsize=14)
ax1.set_xlabel("Time / Phase ($x$)", fontsize=12)
ax1.set_ylabel("Amplitude", fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend()

ax2.plot(x, y_cosine, color='crimson', linewidth=2, label='$\cos(x)$')
ax2.set_title("Cosine Wave Analysis", fontsize=14)
ax2.set_xlabel("Time / Phase ($x$)", fontsize=12)
ax2.set_ylabel("Amplitude", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend()

plt.tight_layout()
plt.show()
```

5. Histogram
Generate a histogram for a dataset of 1000 random numbers sampled from a
normal distribution with mean 0 and standard deviation 1.
o Add bins to the histogram.
o Customize the color and edge color.
o Save the plot as an image file (histogram.png).

```python
np.random.seed(42) 
data = np.random.normal(loc=0, scale=1, size=1000)

fig, ax = plt.subplots(figsize=(8, 6))
ax.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.8)

ax.set_title("Histogram: Normal Distribution ($\mu=0, \sigma=1$)", fontsize=14, fontweight='bold')
ax.set_xlabel("Value", fontsize=12)
ax.set_ylabel("Frequency", fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('histogram.png')
plt.show()
```
